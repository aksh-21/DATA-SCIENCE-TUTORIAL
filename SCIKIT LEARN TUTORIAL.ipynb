### CONVERTING NOW OTHER DATATYPE VALUES INTO NUMERIC VALUES..

```python
#ONE_HOT_ENCODER is used from encoding purposes where it convert the data into 0,1 and we  get a array
from sklearn.preprocessing import OneHotEncoder
# NOW WE WILL IMPORT COLUMN_TRANSFORMER 
# COLUMN_TRANSFORMER IS UESD TO DECLARE A FUNCITON WHICH WITH HELP OF IT ONLY WE CAN APPLY TO SELECTED COLUMNS ONLY
from sklearn.compose import ColumnTransformer
#list the features which i am gonna transform
categorical_features=["Make","Colour","Doors"]
#INSTANTIATE ONE_HOT_ENCODER
ONE_HOT = OneHotEncoder()
#USE OF COLUMNTRANSFORMER
# gonna apply a one_hot over the categorical features transverse column wise and if a any casualty occurs it will pass through
transformer =ColumnTransformer([("one_hot",ONE_HOT,categorical_features)],remainder="passthrough")
# WE Had CREATED A TRANSFORMER OBJECT WHICH WE GONNA #FIT# NOW OVER OUR DATABASE
TRANSFORMED_X=transformer.fit_transform(car_sales_missing)
TRANSFORMED_X
```

## USING SKLEARN LIBRARY
#### WE CAN ALSO FILL THE NAN VALUES IN OUR DATABASE USING THE SKLEARN LIBRARY

```python
# READ A CSV FILE
car_sales_missing = pd.read_csv("car-sales-missing-data.csv")
car_sales_missing
```

```python
# SEE THE WORKING OF ISNA() FUNCTION
car_sales_missing.isna()
```

```python
# HOW TO DROP A ROWS IN DATABASE BASED ON NULL VALUES IN SPECFIC COLUMN
car_sales_missing.dropna(subset=["Price"],inplace= True)
car_sales_missing
```

```python
car_sales_missing.isna().sum()
```

```python
# FILLING THE MISSING VALUES WITH SKLEARN 
# FILLING THE VALUES IS ALSO CALLED IMPUTATION
from sklearn.impute import SimpleImputer
# IMPUTATAION LIBRARY
#simple imputer gonna impute the dataset for us
from sklearn.compose import ColumnTransformer
#FILLING CATEGORICAL VALUES WITH THE "MISSING "KEYWORD AND NUMERICAL WITH MEAN
CAT_IMPUTER=SimpleImputer(strategy="constant",fill_value="missing")
DOOR_IMPUTER=SimpleImputer(strategy="constant",fill_value=4)
NUM_IMPUTER=SimpleImputer(strategy="mean")
#DEFINE DIFF. SET OF COLUMNS WHERE TO APPLY THESE IMPUTERS
CAT_FEATURES=["Make","Colour"]
NUM_FEATURES=["Odometer"]
DOOR_FEATURES=["Doors"]
#create a imputer which when apply to a specific database bring the required changes
IMPUTER=ColumnTransformer([("CAT_TRANS",CAT_IMPUTER,CAT_FEATURES),
                           ("DOOR_TRANS",DOOR_IMPUTER,DOOR_FEATURES),
                           ("NUMERIC_TRANS",NUM_IMPUTER,NUM_FEATURES)],remainder="passthrough")
#FIT TRANSFORM IS A ATTRIBUTE OF THE COLUMN_TRASFORMER
filled_X=IMPUTER.fit_transform(X)
filled_X
# WHEN WE APPLIED THIS SKLEARN TRANSFORMATION TO THE DATABASE WE GET A ARRAY
```

```python
# WE NOW CONVERT THIS ND.ARRAY INTO THE DATAFRAME
CAR_SALES_MISSING= pd.DataFrame(filled_X,columns=["MAKE","COLOUR","DOORS","PRICE"])
CAR_SALES_MISSING
```

```python
#checking the number of NAN values
CAR_SALES_MISSING.isna().sum()
```

## 2.1 PICKING A MACHINE LEARNING MODEL FOR REGRESSION PROBLEM

```python
#importing a built in database
from sklearn.datasets import load_boston
#IMPORT A DATASET FROM A DATASETS OF A SKLEARN(IN BUILT)
#loading a BOSTON dataset
boston = load_boston()
#BOSTON WILL ACTUALLY GIVE A DICTONARY OF DATASETS(WITH KEYS AND VLAUES IN IT)
# CREATING A DATAFRAME FROM DICTONARY..
boston_df=pd.DataFrame(boston["data"],columns=boston["feature_names"])
#HERE DATASET WILL BE VALUES AND ATTRIBUTES NAMES WILL BE FEATURES NAME:
#CREATING A NEW COLUMN IN BOSTON DATA SET
boston_df["target"]=pd.Series(boston["target"])
#boston["target"] in a key in loaded boston dictonary
boston_df.head(10)
```

```python
#HOW MANY SMAPLES!!??
len(boston_df)
```

```python
#LETS TRY RIDGE REGRESSION MODEL OVER THIS DATABASE
from sklearn.linear_model import Ridge
#setup a random seed
np.random.seed(10)
#ASSIGNING FEATURESS TO THE X AND CLASSES/VALUES TO THE Y
# REGRESSION MODELS ARE GENERALLY APPLIED WHEN YOU NEED TO FIND OUT NUMERIC VALUES:
X=boston_df.drop("target",axis=1)
Y=boston_df["target"]
#SPLITTING DATA INTO TRAINING AND TESTING SET
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
#INSTANTIATE THE MODEL 
model=Ridge()
#FIT THE MODEL TO THE TRAINING SET
model.fit(X_train,Y_train)
#check the score of the ridge model on the test data
model.score(X_test,Y_test)
```

### ENSEMBLE METHOD(RANDOMIZED METHOD)

```python
#USING A REGRESSION METHOS FOR ESTIMATING A NUMERIC OUTPUT
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
np.random.seed(10)
#Create a Data
X=boston_df.drop("target",axis=1)
Y=boston_df["target"]
#SPLITTING DATA INTO TRAINING AND TESTING SET
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
#INSTANTIATE THE MODEL 
model=RandomForestRegressor(n_estimators=100)
#FIT THE MODEL TO THE TRAINING SET
model.fit(X_train,Y_train)
#check the score of the ridge model on the test data
model.score(X_test,Y_test)
```
